# Example 1: Validation-Test Split

```python
#Example 1: Logistic-Validation set
#Import necessary libraries:
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# fake data for exercise
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, weights=[0.1, 0.9], random_state=1)

# Split the data using stratified sampling and fixed a random seed (state) for reproduce the results
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)


# Choose a model!
# https://scikit-learn.org/stable/supervised_learning.html
model = LogisticRegression()
#model_ols = LinearRegression()

# Fit the model
model.fit(X_train, y_train)

# Make (0,1) predictions and evaluate the model:
y_pred = model.predict(X_test)

# Make predicted probabilities and evaluate the model:
y_predict_proba = model.predict_proba(X_test)

# Evaluate the model prediction using a particular score:
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# look into the confusion matrix the results
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
(tn, fp, fn, tp)
```

# Example 2: Five-fold cross-validation

```python

#Import necessary libraries:
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error

# generate data
X, y = make_regression(n_samples=1000, n_features=10, noise=0.1, random_state=1)

# ------------------------------------------------------------------------
# Approach 1: <Loop>
# set up the 5-fold
kf = KFold(n_splits=10, random_state=1, shuffle=True)

#Initialize the OLS linear regression model:
model = LinearRegression()

# model fitting and evaluate the 5-fold MSE
mse_list = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

# Report the mean of 5-fold MSE
avg_mse = np.mean(mse_list)
print('Average Mean Squared Error:', avg_mse)

# ------------------------------------------------------------------------
# Approach 2: <cross_validate>
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html#sklearn.model_selection.cross_validate

from sklearn.model_selection import cross_validate
results = cross_validate(model, X, y, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
avg_train_mse = np.mean(-results['train_score'])
avg_test_mse = np.mean(-results['test_score'])
print('Average Train Mean Squared Error:', avg_train_mse)
print('Average Test Mean Squared Error:', avg_test_mse)

# ------------------------------------------------------------------------
# Approach 3: <cross_val_score>
from sklearn.model_selection import cross_val_score
mse_scores = -cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
avg_mse = np.mean(mse_scores)
print('Average Mean Squared Error:', avg_mse)
```

#Example 3: Use many classifiers and performance metrics

```python

# load the packages
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
# Sklearn sample split
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, StratifiedKFold, cross_val_predict
# performace metrics
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import matthews_corrcoef, make_scorer
matthews_corrcoef = make_scorer(matthews_corrcoef)
# Import SkLearn common classifiers #
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier, GradientBoostingClassifier
from sklearn.linear_model import Lasso, LassoCV, SGDClassifier, LinearRegression, LogisticRegression, RidgeCV, RidgeClassifierCV, HuberRegressor, LinearRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

# define classifiers and speeficy the parameters and random seed
classifiers = {
  'OLS':               LinearRegression(),
  "Logistic":          LogisticRegression(),
  "Ridge":             RidgeClassifierCV(),
  'LASSO':             LassoCV(n_jobs= -1, random_state = 1234),
  'Robust':            HuberRegressor(),
   "NNB":              KNeighborsClassifier(10, n_jobs= -1),
   "LinearSVC":        LinearSVC(random_state= 1234, dual = False, max_iter = 2000),
   "SVC":              SVC(probability= True),
   "AdaBoost":         AdaBoostClassifier(random_state= 1234),
   "Bagging":          BaggingClassifier(n_jobs= -1,random_state= 1234),
   "RandomForest":     RandomForestClassifier(n_estimators= 2000,  n_jobs= -1, random_state = 1234),
   "GradientBoosting": GradientBoostingClassifier(random_state= 1234),
   "XGboost":          XGBClassifier(n_jobs= -1,random_state= 1234),
   "MLP":              MLPClassifier()
   }

# define metrics names
metric_names = ["roc_auc",    "roc_auc_05",
                "accuracy_05",
                "f1_05",
                "precision_05",
                "recall_05",      "share"]

# fake data for exercise
X, Y = make_classification(n_samples=1000, n_features=20, n_classes=2, weights=[0.1, 0.9], random_state=1)

# stratified 5-Fold
kfold = 5
skf = StratifiedKFold(n_splits=kfold, shuffle=True, random_state=1234)

# define the  saved stuff after you run the loop
# you can take the mean of five folds (In)
measures_In  = {key: pd.DataFrame(index = metric_names, columns = range(kfold)) for key in classifiers}

# or after each fold, collect the y_predict and append cross fold, then use the fina y_pred and actual y to calc,.the metrics
measures_Out = {key: pd.DataFrame(index = metric_names, columns = range(1)) for key in classifiers}
# for each ML, collect the 5-fold preditions/probs
y_test_collected           = {key: [] for key in classifiers}
y_pred_prob_collected      = {key: [] for key in classifiers}
y_pred_05_collected        = {key: [] for key in classifiers}

# loop for 5-fold
for cv, (train_idx, test_idx) in enumerate(skf.split(X, Y)):
  X_train, y_train = X[train_idx], Y[train_idx]
  X_test, y_test   = X[test_idx], Y[test_idx]

  # loop for models. Note some models don't report the y_pred_prob! (you can check each ML's output using Example 1)
  for clf_name, clf in classifiers.items():
    print(clf_name)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    if clf_name in ["OLS", "LASSO", "Robust", "Ridge" ,"LinearSVC"]:
      y_pred_prob = y_pred
      y_pred = np.where(y_pred > 0.5, 1, 0)
    else:
      y_pred_prob = clf.predict_proba(X_test)[:, 1]

    # record the measures (for each fold results, calc the metrics)
    measures_In[clf_name].loc["roc_auc", cv]       = roc_auc_score(y_test,   y_pred_prob)
    measures_In[clf_name].loc["roc_auc_05", cv]    = roc_auc_score(y_test,   y_pred)
    measures_In[clf_name].loc["accuracy_05", cv]   = accuracy_score(y_test,  y_pred)
    measures_In[clf_name].loc["f1_05", cv]         = f1_score(y_test,        y_pred)
    measures_In[clf_name].loc["precision_05", cv]  = precision_score(y_test, y_pred)
    measures_In[clf_name].loc["recall_05", cv]     = recall_score(y_test,    y_pred)
    measures_In[clf_name].loc["share", cv]         = np.mean(y_pred)
    # append each fold's predicion/actual y
    y_test_collected[clf_name].extend(list(y_test))
    y_pred_prob_collected[clf_name].extend(list(y_pred_prob))
    y_pred_05_collected[clf_name].extend(list(y_pred))

  # after each classifier, use the 'append' y_predict to calc. the metrics
  for clf_name in classifiers:
    measures_Out[clf_name].loc["roc_auc", 0]      = roc_auc_score(y_test_collected[clf_name],   y_pred_prob_collected[clf_name])
    measures_Out[clf_name].loc["roc_auc_05", 0]   = roc_auc_score(y_test_collected[clf_name],   y_pred_05_collected[clf_name])
    measures_Out[clf_name].loc["accuracy_05", 0]  = accuracy_score(y_test_collected[clf_name],  y_pred_05_collected[clf_name])
    measures_Out[clf_name].loc["f1_05", 0]        = f1_score(y_test_collected[clf_name],        y_pred_05_collected[clf_name])
    measures_Out[clf_name].loc["precision_05", 0] = precision_score(y_test_collected[clf_name], y_pred_05_collected[clf_name])
    measures_Out[clf_name].loc["recall_05", 0]    = recall_score(y_test_collected[clf_name],    y_pred_05_collected[clf_name])
    measures_Out[clf_name].loc["share", 0]        = np.mean(y_pred_05_collected[clf_name])
```
