## 2. Bag of Words (BoW)

#### What is Bag of Words (BoW)?
After understanding word frequency, the next step is Bag of Words (BoW) because it transforms word counts into a structured numerical representation.It is a text representation technique widely used in NLP.

How does BoW differ from word frequency?
 - **Word frequency** simply counts how often each word appears in a document.
 - **Bag of Words (BoW)** takes this concept further by converting word frequencies into a vector, making it easier to analyze and compare different texts.

Key Characteristics of BoW:
 - **Transforms Text into Numbers**: Converts words into numerical data for machine learning.
 - **Easy to Use**: Simple and efficient for basic text analysis.
 - **Ignores Word Order**: Focuses on word frequency, not sentence structure.
 - **Large Vocabulary Size**: The vector size depends on the number of unique words.
 - **Sparse Data**: Most texts use only a small part of the total vocabulary, making BoW representations mostly empty.

#### Example:
Suppose we have the following two sentences:
1. "I love machine learning"
2. "Machine learning is fun"

The BoW model will construct a vocabulary like this: ["I", "love", "machine", "learning", "is", "fun"]

Then, each sentence will be transformed into a frequency vector:
1. Sentence 1: [1, 1, 1, 1, 0, 0]   
2. Sentence 2: [0, 0, 1, 1, 1, 1]

 - Some words appear frequently, but are they meaningful?\
 Imagine you have an article about cats. The word "cat" appears 10 times, but words like "the," "is," and "a" appear 30 times.\
 Does that mean "the" is more important than "cat"? Obviously not.
 - Word frequency helps us identify significant words in a document because frequently occurring words might be the main topic. \

## 3. Term Frequency-Inverse Document Frequency

### What is Term Frequency-Inverse Document Frequency (TF-IDF) ?
TF-IDF is a statistical measure used in NLP to evaluate the importance of a word in a document relative to a collection of documents. It is an improvement over the Bag of Words (BoW) model because it not only considers **word frequency** but also **reduces the weight of common words** that appear in many documents.

## **TF and IDF:**
### **Term Frequency (TF):**  
Measures how often a word appears in a document. The basic formula is:  

$$
TF = \frac{\text{Number of times the word appears in the document}}{\text{Total number of words in the document}}
$$  

Words that appear more frequently in a document will have a higher TF score.

---

### **Inverse Document Frequency (IDF):**  
Reduces the weight of common words by measuring how important a word is across the entire corpus. The formula is:  

$$
IDF = \log \left( \frac{\text{Total number of documents}}{\text{Number of documents containing the word}} \right)
$$

- If a word appears in many documents, its IDF score will be low.  
- If a word appears in only a few documents, its IDF score will be high.

---

### **TF-IDF Score:**  

$$
TF\text{-}IDF = TF \times IDF
$$

- Words with high TF-IDF scores are more unique and relevant to a document.  
- Common words (e.g., "the", "is", "and") will have lower scores due to their low IDF values.
---


#### Example 1:
Consider the following three documents:
1. Document 1: "I love machine learning."
2. Document 2: "Machine is fun."
3. Document 3: "I love deep learning."

- For "Machine"

    - **Step 1:** Calculate TF

        For "machine" in Document 1:  
$$
        TF = \frac{1}{4} = 0.25
$$  
        For "machine" in Document 2:  
        $$
        TF = \frac{1}{3} = 0.33
        $$  

    - **Step 2:** Calculate IDF

        Since "machine" appears in 2 out of 3 documents:  
        $$ IDF = \log\left(\frac{3}{2}\right) = 0.176 $$  

    - **Step 3:** Calculate TF-IDF Score

        For "machine" in Document 1:  
        $$ TF-IDF = 0.25 \times 0.176 = 0.044 $$  
        For "machine" in Document 2:  
        $$ TF-IDF = 0.33 \times 0.176 = 0.0581 $$  

- For "deep"

    - **Step 1:** Calculate TF  
        For "deep" in Document 3:  
        $$ TF = \frac{1}{4} = 0.25 $$  

    - **Step 2:** Calculate IDF  
        Since "deep" appears in 1 out of 3 documents:  
        $$ IDF = \log\left(\frac{3}{1}\right) = 0.477 $$  

    - **Step 3:** Calculate TF-IDF Score  
        For "deep" in Document 3:  
        $$ TF-IDF = 0.25 \times 0.477 = 0.11925 $$  

Since "deep" appears only in Document 3, its IDF value will be much higher, making its TF-IDF score larger compared to common words like "machine" and "learning."

### TF-IDF Function

Here is a reusable TF-IDF function that can be called directly:

````python
from sklearn.feature_extraction.text import TfidfVectorizer

def compute_tfidf(corpus, max_features=None):
    """
    Compute TF-IDF
    corpus: list of strings, each entry is a document
    max_features: optional, limits the number of features to keep
    Returns: (tfidf_matrix, feature_names)
    """
    vectorizer = TfidfVectorizer(max_features=max_features)
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = vectorizer.get_feature_names_out()
    return tfidf_matrix, feature_names
````

````python
import pandas as pd

# Example Corpus
corpus = [
    "The stock market crashed yesterday due to economic uncertainty",
    "Central bank raised interest rates to combat inflation",
    "Investors are optimistic about the stock market recovery",
    "Higher interest rates affect consumer spending and borrowing"
]

# TF-IDF Calculation
tfidf_matrix, feature_names = compute_tfidf(corpus)

# Convert the TF-IDF result to a DataFrame for readability
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
df_tfidf
````

````python
# Display each document's TF-IDF vector
for i in range(len(corpus)):
    vector = df_tfidf.iloc[i].values  # TF-IDF vector for document i
    print(f"Document {i} TF-IDF vector:\n{vector}\n")
````

Input / Output Dimensions:

 - Input: A list of strings.
 - Output: A sparse matrix of shape (n_docs, n_words) containing the TF-IDF scores, and a list of the corresponding feature names (words).

#### The Limitation of TF-IDF in Time-Sensitive Data:
 - Keyword importance changes over time
   - Some terms become more common due to technological advancements or shifts in research focus.

 - Example: The word "green"
   1. Before 2010: Rarely used in patents.
   2. After 2010: Widely used due to new green technologies.

 - Issue with TF-IDF
    - TF-IDF calculates importance from a modern perspective (2025) across all historical data.
    - Since "green" appears frequently after 2010, its IDF score decreases.
    - This makes "green" seem less important in patents before 2010, even though it was actually rare and significant at the time.

TF-IDF_green.png

 - Solution: TF-BIDF
    - TF-BIDF evaluates keyword importance based on historical context at each time period.
    - Instead of considering all documents together, it looks only at data up to that point in time.
    - This prevents terms from being unfairly downweighted just because they became more common later.

## 4. Term Frequency-Backward Inverse Document Frequency

### What is Term Frequency-Backward Inverse Document Frequency (TF-BIDF)?
TF-BIDF is an extension of the traditional TF-IDF method, specifically designed for time-sensitive data or sequential documents.

#### Key Characteristics of TF-BIDF:
 - Emphasizes terms that are more unique within historical context.
 - Suitable for analyzing temporal data, such as financial reports, historical news, or evolving public opinions.
 - Reduces the influence of frequently mentioned words in earlier time periods.

**Step 1: Term Frequency (TF)**
 - The **TF** component in TF-BIDF is identical to the one in TF-IDF.
 - It measures how frequently a term appears in a document:

    $$
    TF = \frac{\text{Number of times the word appears in the document}}{\text{Total number of words in the document}}
    $$

**Step 2: Backward Inverse Document Frequency (BIDF)**
 - **BIDF** is similar to IDF but considers only the documents that occurred **before or up to the given time point**.

    $$
    BIDF = \log \left( \frac{\text{Total number of past documents}}{\text{Number of past documents containing the word}} \right)
    $$

 - **Key Difference from IDF:**\
    In **BIDF**, only **past documents** are counted, making it ideal for analyzing evolving trends, news timelines, or sequential text data.

**Step 3: TF-BIDF Score**
 - The overall TF-BIDF score is calculated as follows:

  $$
  TF\text{-}BIDF = TF \times BIDF
  $$
---

#### Example: News Articles Over Time
Suppose we are analyzing the term "climate change" in news articles over multiple years:

| Year | Document Text                                    |
|:-----|:-------------------------------------------------|
| 2018 | "Global warming is impacting bad effects."       |
| 2019 | "New climate change policies are being discussed."|
| 2020 | "Climate change has become a major concern."     |

- **Step 1:**  
  Calculate TF for "climate" in the 2020 document:
  
  $$ TF = \frac{1}{7} = 0.14 $$

- **Step 2:** Calculate BIDF for "climate" in 2020:
    - Up to 2020, there are 2 documents in the past.
    - The term "climate" appears in 1 of the past documents (2019).
  
  $$ BIDF = \log\left(\frac{2}{1}\right) = 0.301 $$

- **Step 3:**  
  Calculate TF-BIDF:
  
  $$ TF-BIDF = TF \times BIDF = 0.14 \times 0.301 = 0.042 $$  
---

Conclusion:\
TF-BIDF is particularly useful for analyzing textual trends over time, such as:
 - Financial trend analysis
 - Tracking political discourse
 - Monitoring scientific developments across years

## Example: Financial News by Year

```test
Year  text

2010  China's economy grew strongly, with exports as the main driver.

2015  the global economy slowed down, and China shifted to domestic demand to sustain growth.

2020  the COVID-19 pandemic disrupted global supply chains, causing severe economic damage.

2023  China's economy gradually recovered, with tech industries and green energy becoming the new focus.
```

| Method      |  Description                                                                      |
|-------------|----------------------------------------------------------------------------------|
| Count       | Counts how many times a word appears in a document.                             |
| TF          | Term Frequency = word count / total words in the document.                      |
| TF-IDF      | Highlights words that are frequent in one document but rare across others.      |
| TF-BIDF     | Like TF-IDF, but emphasizes time-based rarity.              |


````python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
import numpy as np

docs = [
    "China's economy grew strongly, with exports as the main driver.",
    "the global economy slowed down, and China shifted to domestic demand to sustain growth.",
    "the COVID-19 pandemic disrupted global supply chains, causing severe economic damage.",
    "China's economy gradually recovered, with tech industries and green energy becoming the new focus."
]
years = [2010, 2015, 2020, 2023]
````

#### Word Frequency (Count)

````python
# Count Vectorizer
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(docs)

# Convert to DataFrame
count_df = pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
count_df
````

Convert Count DataFrame to Vectors:

````python
# Convert each row of the count DataFrame into a vector
count_vectors = [row.values for _, row in count_df.iterrows()]

# Print as individual vectors
for i, vec in enumerate(count_vectors):
    print(f"Document {i+1} - Count Vector:\n{vec}\n")
````

#### Bag of Words (BoW)

````python
import numpy as np

# BoW = count / total terms in doc
BoW_matrix = count_matrix.astype(float)
for i in range(BoW_matrix.shape[0]):
    row_sum = BoW_matrix[i].sum()
    if row_sum != 0:
        BoW_matrix[i] /= row_sum

# Convert to DataFrame
BoW_df = pd.DataFrame(BoW_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
BoW_df
````

Convert BoW (TF) DataFrame to Vectors:

````python
# Convert each row of the BoW (TF) DataFrame into a vector
bow_vectors = [row.values for _, row in BoW_df.iterrows()]

# Print as individual vectors
for i, vec in enumerate(bow_vectors):
    print(f"Document {i+1} - BoW (TF) Vector:\n{vec}\n")
````

#### TF-IDF

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(docs)

# Convert to DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df
````

Convert TF-IDF DataFrame to Vectors:

````python
# Convert each row of the TF-IDF DataFrame into a vector
tfidf_vectors = [row.values for _, row in tfidf_df.iterrows()]

# Print as individual vectors
for i, vec in enumerate(tfidf_vectors):
    print(f"Document {i+1} - TF-IDF Vector:\n{vec}\n")
````

#### TF-BIDF

````python
from collections import defaultdict
from math import log

# Create a set of all years per word
word_years = defaultdict(set)
for doc, year in zip(docs, years):
    words = doc.lower().split()
    for word in set(words):
        word_years[word].add(year)

# Calculate BIDF (log(total years / # of years word appears)
total_years = len(set(years))
bidf = {word: log(total_years / len(year_set)) for word, year_set in word_years.items()}

# Compute TF-BIDF manually
tf_bidf_matrix = []
for doc in docs:
    words = doc.lower().split()
    word_count = defaultdict(int)
    for word in words:
        word_count[word] += 1
    total_words = len(words)
    tf_bidf_row = {}
    for word in word_count:
        tf = word_count[word] / total_words
        tf_bidf_row[word] = tf * bidf.get(word, 0)
    tf_bidf_matrix.append(tf_bidf_row)

# Convert to DataFrame
tf_bidf_df = pd.DataFrame(tf_bidf_matrix).fillna(0)
tf_bidf_df
````

Convert TF-BIDF DataFrame to Vectors:

````python
# Convert each row of the TF-BIDF DataFrame into a vector
tfbidf_vectors = [row.values for _, row in tf_bidf_df.iterrows()]

# Print as individual vectors
for i, vec in enumerate(tfbidf_vectors):
    print(f"Document {i+1} - TF-BIDF Vector:\n{vec}\n")
````
