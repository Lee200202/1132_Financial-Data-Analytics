# **K-mean Cluster**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Create data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=2.0, random_state=1234)

# Choose random points as initial centroids
np.random.seed(42) # for reproducibility
initial_centroids = X[np.random.choice(X.shape[0], size=4, replace=False)]

def find_closest_centroids(X, centroids):
    distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
    return np.argmin(distances, axis=0)

def compute_centroids(X, idx, k):
    centroids = np.zeros((k, X.shape[1]))
    for i in range(k):
        centroids[i, :] = np.mean(X[idx == i, :], axis=0)
    return centroids

def plot_data(X, centroids, idx=None):
    """
    Visualize the data points and centroids.
    """
    plt.scatter(X[:, 0], X[:, 1], c=idx, cmap='viridis', s=30, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X')
    plt.title("Data points and Centroids")
    plt.show()

# Initialize values
k = 4
max_iters = 5
centroids = initial_centroids

# Run k-means algorithm
for i in range(max_iters):
    # For each point, find the centroid it is closest to
    idx = find_closest_centroids(X, centroids)

    # Plot the data and centroids
    plot_data(X, centroids, idx)

    # Compute new centroids
    centroids = compute_centroids(X, idx, k)
```

```python
scaler = StandardScaler()
nci_scaled = scaler.fit_transform(nci_data)
```

We now perform hierarchical clustering of the cell lines in the NCI60 data using complete, single, and average linkage. Once again, the goal is to find out whether or not the observations cluster into distinct types of cancer. Euclidean distance is used as the dissimilarity measure. We first write a short function to produce the three dendrograms.


```python
HClust = AgglomerativeClustering
def plot_nci(linkage, ax, cut=-np.inf):
    cargs = {'above_threshold_color':'black',
             'color_threshold':cut}
    hc = HClust(n_clusters=None,
                distance_threshold=0,
                linkage=linkage.lower()).fit(nci_scaled)
    linkage_ = compute_linkage(hc)
    dendrogram(linkage_,
               ax=ax,
               labels=np.asarray(nci_labs),
               leaf_font_size=10,
               **cargs)
    ax.set_title('%s Linkage' % linkage)
    return hc
```

# Application 1: Cluster the style of mutual funds

```python
import numpy as np
from sklearn.cluster import KMeans

def generate_synthetic_data_tn_with_true_groups(T=1000, N=10):
    """
    Generate synthetic data shaped (T x N) with true style groups.
    Each column represents a fund and each row a time point.
    True style groups:
      Group 0: loc = 0.0, scale = 0.01
      Group 1: loc = 0.001, scale = 0.02
      Group 2: loc = -0.001, scale = 0.015
    Returns:
      X: np.ndarray of shape (T, N)
      true_groups: np.ndarray of shape (N,) with values 0, 1, or 2
    """
    np.random.seed(42)
    true_groups = np.random.choice([0, 1, 2], size=N)
    X = np.zeros((T, N))
    for j in range(N):
        if true_groups[j] == 0:
            X[:, j] = np.random.normal(loc=0.0, scale=0.5, size=T)
        elif true_groups[j] == 1:
            X[:, j] = np.random.normal(loc=1, scale=0.1, size=T)
        else:
            X[:, j] = np.random.normal(loc=-1, scale=0.3, size=T)
    return X, true_groups

def gsc_clustering_tn_improved(X_tn, K, max_iter=100, tol=1e-4, random_state=42, min_std=1e-3):
    """
    Improved GSC clustering for (T x N) data with a minimum variance floor and
    global normalization. Clusters N funds (columns) based on their T-dimensional time series.

    Parameters
    ----------
    X_tn : np.ndarray
        Shape (T, N). Each column is a fund.
    K : int
        Number of clusters.
    max_iter : int
        Maximum iterations.
    tol : float
        Convergence threshold.
    random_state : int
        Seed.
    min_std : float
        Minimum standard deviation to enforce (avoids near-zero issues).

    Returns
    -------
    best_assignments : np.ndarray
        Best cluster assignments (N,).
    best_means : np.ndarray
        Final cluster means (K, T) in normalized space.
    best_stds : np.ndarray
        Final cluster standard deviations (K,).
    best_obj : float
        Objective (sum of weighted squared distances) of the best run.
    """
    np.random.seed(random_state)
    # Transpose to get (N x T): each fund is a row.
    X_nt = X_tn.T
    N, T = X_nt.shape

    # Global normalization (all funds, all time points)
    overall_mean = X_nt.mean()
    overall_std = X_nt.std() + 1e-9
    X_norm = (X_nt - overall_mean) / overall_std

    best_obj = np.inf
    best_assignments, best_means, best_stds = None, None, None

    n_restarts = 10  # Use multiple random initializations
    for restart in range(n_restarts):
        # Randomly initialize cluster centers from the funds
        init_indices = np.random.choice(N, K, replace=False)
        cluster_means = X_norm[init_indices, :].copy()
        cluster_stds = np.ones(K)
        assignments = np.zeros(N, dtype=int)
        obj_value = np.inf

        for iteration in range(max_iter):
            old_means = cluster_means.copy()
            # Assignment step: compute variance-weighted distances
            distances = np.zeros((N, K))
            for k in range(K):
                diff = X_norm - cluster_means[k]  # shape (N, T)
                sq_dist = np.sum(diff**2, axis=1)
                distances[:, k] = sq_dist / ((cluster_stds[k]**2) + 1e-9)
            new_assignments = np.argmin(distances, axis=1)
            obj_new = np.sum([distances[i, new_assignments[i]] for i in range(N)])
            # Check if assignments are unchanged (or objective stabilizes)
            if np.array_equal(new_assignments, assignments) and abs(obj_value - obj_new) < tol:
                assignments = new_assignments
                obj_value = obj_new
                break
            assignments = new_assignments
            obj_value = obj_new

            # Update step: recalc cluster means and variances
            for k in range(K):
                members = X_norm[assignments == k]
                if len(members) > 0:
                    cluster_means[k] = members.mean(axis=0)
                    var_k = np.mean(np.sum((members - cluster_means[k])**2, axis=1))
                    cluster_stds[k] = max(np.sqrt(var_k + 1e-9), min_std)
                else:
                    # Reinitialize an empty cluster
                    rand_idx = np.random.choice(N)
                    cluster_means[k] = X_norm[rand_idx]
                    cluster_stds[k] = 1.0

            if np.sqrt(np.sum((cluster_means - old_means)**2)) < tol:
                break

        if obj_value < best_obj:
            best_obj = obj_value
            best_assignments = assignments.copy()
            best_means = cluster_means.copy()
            best_stds = cluster_stds.copy()

    return best_assignments, best_means, best_stds, best_obj

if __name__ == "__main__":
    # Parameters
    T = 1000  # number of time points (days)
    N = 10    # number of funds
    K = 3     # desired number of clusters

    # Generate synthetic data (T x N) and the true style groups
    X_tn, true_groups = generate_synthetic_data_tn_with_true_groups(T=T, N=N)

    # --- Run Improved GSC Clustering ---
    gsc_assignments, gsc_means, gsc_stds, gsc_obj = gsc_clustering_tn_improved(
        X_tn, K=K, max_iter=100, tol=1e-4, random_state=123
    )

    # --- Run K-means Clustering ---
    # For K-means we need the funds as rows: shape (N, T)
    X_nt = X_tn.T
    # We apply global normalization as before for a fair comparison:
    overall_mean = X_nt.mean()
    overall_std = X_nt.std() + 1e-9
    X_norm = (X_nt - overall_mean) / overall_std
    kmeans = KMeans(n_clusters=K, random_state=123)
    kmeans.fit(X_norm)
    kmeans_assignments = kmeans.labels_

    # --- Print Comparison ---
    print("Comparison of Clustering Results:")
    print("---------------------------------")
    print("Fund    True Group    GSC Cluster    KMeans Cluster")
    for i in range(N):
        print(f"{i:4d}    {true_groups[i]:10d}    {gsc_assignments[i]:11d}    {kmeans_assignments[i]:14d}")

    print("\nGSC Objective Value:", gsc_obj)
    print("\nGSC Cluster Standard Deviations (normalized space):", gsc_stds)
```
